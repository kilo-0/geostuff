```python
from geopandas import GeoSeries, GeoDataFrame as gdf
from shapely.geometry import Polygon, Point, LineString, box, Polygon
import geopandas as gpd
import shapely._geometry
import pandas as pd
import folium
from shapely.geometry import box
import math

def divide_box_into_squares(xmin, xmax, ymin, ymax, square_miles):
    # Define the size of each square in degrees (approximation)
    square_size = math.sqrt(square_miles) / 69  # 1 degree is approximately 69 miles

    # Create the main box
    main_box = box(xmin, ymin, xmax, ymax)

    # Calculate the number of squares along x and y axes
    num_squares_x = math.ceil((xmax - xmin) / square_size)
    num_squares_y = math.ceil((ymax - ymin) / square_size)

    # Generate the smaller squares
    squares = []
    for i in range(num_squares_x):
        for j in range(num_squares_y):
            x1 = xmin + i * square_size
            x2 = min(x1 + square_size, xmax)
            y1 = ymin + j * square_size
            y2 = min(y1 + square_size, ymax)
            squares.append(box(x1, y1, x2, y2))

    return squares

# Define the main box geometry (xmin, xmax, ymin, ymax)
#xmin, xmax, ymin, ymax = -125.0, -66.5, 24.0, 49.5

# Divide the box into equal 1000 square mile boxes
squares = divide_box_into_squares(xmin, xmax, ymin, ymax, 1000)

# Print the number of smaller boxes created
#print(f"Number of smaller boxes created: {len(squares)}")
# Divide the box into equal 1000 square mile boxes
squares = divide_box_into_squares(xmin, xmax, ymin, ymax, 1000)

# Create a pandas dataframe with each smaller box geometry in WKT (BOX())
df = pd.DataFrame({'geometry': [square.wkt for square in squares]})

# Display the dataframe
df
```

```python
from shapely import wkt
# Create a base map
m = folium.Map(location=[(ymin + ymax) / 2, (xmin + xmax) / 2], zoom_start=4)

# Add each box to the map
for geom in df['geometry']:
    shape = wkt.loads(geom)
    folium.GeoJson(shape).add_to(m)

# Save the map as an HTML file
m.save("map.html")

# Display the map
m
```
